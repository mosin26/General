```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
data = pd.read_csv('task1.csv')
```

```python
data.head()
```

```python
print(data.shape)
```

```python
x = np.array(data['x'])
y = np.array(data['y'])
```

```python
plt.scatter(x, y, s=10)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x_without_2 = np.delete(x, np.where((x<50) & (y>80)))
y_without_2 = np.delete(y, np.where((x<50) & (y>80)))

x_without_3 = np.delete(x_without_2, np.where(y_without_2<25))
y_without_3 = np.delete(y_without_2, np.where(y_without_2<25))
```

```python
plt.scatter(x, y, s=10)
plt.scatter(x_without_2, y_without_2, s=10)
plt.scatter(x_without_3, y_without_3, s=10)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
z_1 = np.poly1d(np.polyfit(x,y,2))
z_2 = np.poly1d(np.polyfit(x_without_2,y_without_2,2))
z_3 = np.poly1d(np.polyfit(x_without_3,y_without_3,2))
```

```python
r = range(9,101)

plt.plot(r, [z_1(i) for i in r], linewidth=2, label='full sample')
plt.plot(r, [z_2(i) for i in r], linewidth=2, label='without 2 outliers')
plt.plot(r, [z_3(i) for i in r], linewidth=2, label='without 3 outliers')
plt.scatter(x, y, s=10, c='k')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
